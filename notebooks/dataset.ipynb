# Conjunto de Datos del Sector Agrícola en México

Para el desarrollo de este proyecto se hace uso del registro histórico del Servicio de Información Agroalimentaria y Pesquera (SIAP) que comprende la producción agrícola de todos los municipios de México entre los años 2010 y 2024. Este conjunto de datos es la fuente gubernamental oficial que registra la estadística agrícola en México, lo que garantiza que el problema analizado se basa en la realidad operativa del país.

Además, cuenta con el registro de Área Sembrada, Área Siniestrada y variables de contexto crítico como la Modalidad (riego/temporal), el Cultivo y la ubicación geográfica hasta nivel Municipio.

Al abarcar 14 años continuos de registros (2010-2024), proporciona el volumen histórico necesario (más de medio millón de registros) para que los algoritmos de minería de datos puedan encontrar patrones robustos a través del tiempo, capturando las variaciones climáticas de la última década a un nivel de detalle municipal.

## Tamaño del Conjunto de Datos

El archivo utilizado es un volumen de información masivo y sumamente detallado:

*   **Dimensión temporal:** 14 años de registros históricos.
*   **Número de registros (filas):** 541,840 registros individuales.
*   **Granularidad:** A nivel municipal, desglosado por ciclo, cultivo y modalidad hídrica.

Esta magnitud asegura la disponibilidad de datos suficientes para que los algoritmos de minería de datos identifiquen patrones robustos, evitando problemas de escasez o sobreajuste al analizar dinámicas temporales y regionales.

## Descripción de las Variables

Cada fila del dataset representa el resultado de una cosecha para un municipio y cultivo específico en un año determinado. Las variables se dividen en cuatro grandes categorías que ofrecen un panorama holístico del fenómeno agrícola:

### Variables Temporales y Geográficas
Permiten ubicar la siembra en el tiempo y el espacio geográfico:

*   **`Anio`**: El año en que se llevó a cabo el registro de la cosecha (2010 - 2024).

*   **`Nomestado`** y **`Nommunicipio`**: El Estado y el Municipio exacto donde se llevó a cabo la actividad agrícola.

*   **`Nomddr`** y **`Nomcader`**: Nombre del Distrito de Desarrollo Rural (DDR) y del Centro de Apoyo al Desarrollo Rural (CADER). Son demarcaciones administrativas oficiales utilizadas por el SIAP, útiles para análisis regionales más amplios que un municipio.

### Variables de las Condiciones de Cultivo
Describen las características biológicas y técnicas de la siembra:

*   **`Nomcultivo`**: El tipo específico de producto sembrado (ej. Maíz grano, Frijol, Avena forrajera, Tomate rojo, Aguacate).

*   **`Nomcicloproductivo`**: La temporada del año en la que se realizó el cultivo, dividida en: *Otoño-Invierno*, *Primavera-Verano*, o *Perennes* (cultivos de ciclo continuo como árboles frutales).

*   **`Nommodalidad`**: Representa la dependencia hídrica del cultivo. *Riego* indica que cuenta con infraestructura artificial para abastecimiento de agua, mientras que *Temporal* indica una dependencia total del ciclo natural de lluvias.

### Variables de Superficie Agrícola
Miden físicamente las hectáreas de tierra utilizadas y sus resultados de supervivencia:

*   **`Sembrada`**: La superficie total (en hectáreas) que los agricultores plantaron al inicio del ciclo agrícola.

*   **`Cosechada`**: La superficie total (en hectáreas) de la cual se logró extraer producto exitosamente al finalizar el ciclo.

*   **`Siniestrada`**: La superficie total (en hectáreas) donde la siembra se perdió o murió prematuramente, comúnmente debido a desastres climáticos (sequías, heladas, inundaciones) o plagas biológicas. Es el núcleo principal de nuestro análisis de riesgo climático.

### Variables de Producción y Económicas
Evalúan el éxito material y financiero de la cosecha:

*   **`Volumenproduccion`**: La cantidad física total del producto agrícola que se logró extraer, medida generalmente en toneladas.

*   **`Rendimiento`**: Eficiencia productiva calculada como las toneladas extraídas por cada hectárea cosechada.

*   **`Preciomediorural`**: El precio promedio (en pesos mexicanos) al que se le pagó al productor por tonelada del cultivo.

*   **`Valorproduccion`**: El valor económico total generado por la cosecha en esa región (calculado como Volumen de Producción × Precio Medio Rural).
