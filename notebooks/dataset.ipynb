# Conjunto de Datos del Sector Agrícola en México

Los datos abiertos en estadística de producción agrícola son proporcionados por la Dirección General del Servicio de Información Agroalimentaria y Pesquera ([https://nube.agricultura.gob.mx/datosAbiertos/Agricola.php](https://nube.agricultura.gob.mx/datosAbiertos/Agricola.php)). Se recopilaron los datos correspondientes al Nivel Municipal desde el año 2010 hasta el año 2014, donde cada fila representa las estadísticas de producción agrícola para una combinación específica de año, ubicación, ciclo, modalidad de agua y cultivo.

## Descripción de las Variables

### Clasificación de Variables por Categoría Temática

Las 24 variables del dataset pueden agruparse en cinco categorías temáticas según la dimensión del sistema agrícola que representan. Esta clasificación facilita la comprensión de la estructura del dataset.

**Variables de Ubicación Geográfica**

Describen la localización territorial de cada registro, desde el nivel estatal hasta el municipal.

- Isestado / Nomestado
- Idddr / Nomddr
- Idcader / Nomcader
- Idmunicipio / Nommunicipio

**Variables de Contexto Productivo**

Describen las condiciones bajo las cuales se realizó cada siembra: cuándo, cómo y qué se sembró.

- Anio
- Idciclo / Nomcicloproductivo
- Idmodalidad / Nommodalidad
- Idcultivo / Nomcultivo
- Idunidadmedida / Nomunidad

**Variables de Producción Agrícola**

Cuantifican los resultados físicos de cada ciclo productivo en términos de superficie y volumen.

- Sembrada
- Cosechada
- Siniestrada
- Volumenproduccion
- Rendimiento

**Variables Económicas**

Capturan el valor monetario de la actividad agrícola registrada.

- Precio
- Valorproduccion

Estas variables reflejan la dimensión económica del sistema agrícola y son relevantes para cuantificar el impacto financiero de los siniestros.


## Glosario de Terminología Agrícola

Para interpretar correctamente los resultados del análisis es necesario comprender la terminología específica del sector agrícola mexicano utilizada en el dataset y a lo largo de este reporte.

---

**Área sembrada**
Superficie agrícola en la que se depositó la semilla de un cultivo, previa preparación del suelo. Se mide en hectáreas y representa el punto de partida de cada ciclo productivo. No toda la superficie sembrada llega a producir: una parte puede cosecharse exitosamente y otra puede perderse por siniestro.

**Área cosechada**
Fracción del área sembrada de la cual se obtuvo producción efectiva del cultivo. En condiciones normales, la mayor parte del área sembrada se cosecha; la diferencia entre ambas refleja las pérdidas ocurridas durante el ciclo.

**Área siniestrada**
Superficie de la cual no se obtuvo producción debido a que el cultivo se perdió totalmente por alguna causa externa, ya sea climática (sequía, helada, granizo, inundación), biológica (plaga, enfermedad) u otra. Esta variable aplica únicamente a cultivos anuales, ya que los cultivos perennes (frutales, agave, caña) pueden recuperarse en el ciclo siguiente sin necesidad de resembrar.

**Siniestro agrícola**
Evento que provoca la pérdida total o parcial de un cultivo antes de su cosecha. Los siniestros pueden ser de origen climático (el más común en México), biológico o accidental. En el contexto de seguros agrícolas, el siniestro es el evento que activa el pago de indemnización al productor.

**Modalidad hídrica**
Clasificación de una siembra según su fuente de agua. Existen dos modalidades en el dataset:

- **Temporal:** la siembra depende exclusivamente de las precipitaciones naturales. Es la modalidad predominante en México en términos de superficie y la más vulnerable a variaciones climáticas.

- **Riego:** la siembra cuenta con infraestructura hídrica que permite suministrar agua de forma controlada (canales, presas, pozos, sistemas de goteo). Ofrece mayor resiliencia ante sequías pero no elimina completamente el riesgo de siniestro por heladas, plagas u otros factores.

**Ciclo agrícola**
Periodo del año durante el cual se desarrolla un cultivo desde la siembra hasta la cosecha. El dataset registra tres ciclos:

- **Primavera-Verano:** el ciclo principal, que aprovecha las lluvias de mayo a octubre. Concentra la mayor parte de la agricultura de temporal y es el más expuesto al riesgo climático estacional.

- **Otoño-Invierno:** ciclo secundario que se desarrolla entre octubre y marzo, predominante en zonas de riego del noroeste del país donde las temperaturas lo permiten.

- **Perennes:** cultivos que no se ciñen a un ciclo anual específico porque su ciclo de vida supera el año (frutales, agave, caña de azúcar, café). Se registran de forma continua independientemente de la estación.

**Distrito de Desarrollo Rural (DDR)**
Unidad administrativa de la Secretaría de Agricultura, Ganadería, Desarrollo Rural, Pesca y Alimentación (SADER) que agrupa varios municipios con características productivas similares para fines de planificación y apoyo técnico. México cuenta con 191 DDR distribuidos en todo el territorio nacional.

**Centro de Apoyo al Desarrollo Rural (CADER)**
Subdivisión del Distrito de Desarrollo Rural que opera a nivel local, brindando asistencia técnica y administrativa directamente a los productores. Cada DDR está compuesto por varios CADER.

**Precio medio rural**
Precio promedio al que el productor vende su cosecha en el punto de origen, antes de cualquier proceso de transformación o comercialización. Se expresa en pesos por tonelada (o por la unidad de medida correspondiente al cultivo) y refleja el valor de mercado en condiciones normales de la región productora.

**Rendimiento agrícola**
Producción obtenida por unidad de superficie cosechada, expresada generalmente en toneladas por hectárea. Es un indicador de la eficiencia productiva del cultivo en las condiciones específicas de ese año, región y modalidad. Valores bajos de rendimiento pueden indicar condiciones climáticas adversas, aunque no necesariamente implican siniestro total.

**Seguridad alimentaria**
Condición en la que todas las personas tienen acceso físico, social y económico a alimentos suficientes, seguros y nutritivos que satisfacen sus necesidades. La pérdida de cultivos por siniestro amenaza directamente la seguridad alimentaria al reducir la disponibilidad de alimentos básicos y elevar sus precios.

**Milpa**
Sistema agrícola tradicional mexicano basado en el cultivo simultáneo de maíz, frijol y calabaza en la misma parcela. Representa la base de la agricultura campesina de subsistencia en México y explica la alta correlación entre registros de maíz y frijol observada en el dataset.
