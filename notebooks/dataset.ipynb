# Conjunto de Datos del Sector Agrícola en México

Los datos abiertos en estadística de producción agrícola son proporcionados por la Dirección General del Servicio de Información Agroalimentaria y Pesquera [https://nube.agricultura.gob.mx/datosAbiertos/Agricola.php](https://nube.agricultura.gob.mx/datosAbiertos/Agricola.php). Se recopilaron los datos correspondientes al Nivel Municipal desde el año 2010 hasta el año 2014, esto nos da un volumen de 541,840 registros, donde cada fila representa las estadísticas de producción agrícola para una combinación específica de año, ubicación, ciclo, modalidad de agua y cultivo.

## Descripción de las Variables

Cada fila del dataset representa el resultado de una cosecha para un municipio y cultivo específico en un año determinado.
